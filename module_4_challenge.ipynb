## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'line_subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def calculate_shipping_price(row):
   unit_weight = row['unit_weight']
   qty = row['qty']
   if unit_weight * qty > 50:
      return (unit_weight * qty) * 7 
   else:
      return (unit_weight * qty) * 10

# creating a column for total weight:

df['total_weight'] = df['unit_weight'] * df['qty']
    
# # next creating and displaying a 'shipping_price' column     
df['shipping_price'] = df.apply(calculate_shipping_price, axis=1)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price' ]].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['line_subtotal'] = df['unit_price'] * df['qty']

# Next creating a new column that calculates the total price with subtotal, 
# shipping price, and sales tax added on top 
df['line_price'] = ((df['line_subtotal'] + df['shipping_price']) * 1.0925).round(2)
df[['line_subtotal', 'shipping_price', 'line_price']].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()

```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_profit'] = (df['line_price'] - df['line_cost']).round(2)
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# Group the data by client_id and sum the total_price column
def total_spent_by_client(client_id, df):
    return df.loc[df.client_id == client_id, 'line_price'].sum()


[print(f"{client}: ${total_spent_by_client(client, df):,.2f}") for client in top_five]
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
columns_of_interest = ['client_id', 'qty', 'shipping_price', 'line_price', 'line_cost', 'line_profit']
top_five_spend_df = df.query('client_id in [33615, 66037, 46820, 38378, 24741]')[columns_of_interest]\
    .groupby('client_id', as_index=False)\
        .agg(sum)
        
top_five_spend_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
renamed_top_five_spend_df = top_five_spend_df.rename(columns={"client_id": "Client ID", 
 "qty": "Total Units", 
 "shipping_price": "Shipping Cost", 
 "line_price": "Total Revenue",
 "line_cost": "Total Cost", 
 "line_profit": "Total Profit"})

renamed_top_five_spend_df['Shipping Cost'] = renamed_top_five_spend_df['Shipping Cost'].apply(lambda x: '${:,.2f}M'.format(x/1000000))
renamed_top_five_spend_df['Total Revenue'] = renamed_top_five_spend_df['Total Revenue'].apply(lambda x: '${:,.2f}M'.format(x/1000000))
renamed_top_five_spend_df['Total Cost'] = renamed_top_five_spend_df['Total Cost'].apply(lambda x: '${:,.2f}M'.format(x/1000000))
renamed_top_five_spend_df['Total Profit'] = renamed_top_five_spend_df['Total Profit'].apply(lambda x: '${:,.2f}M'.format(x/1000000))

renamed_top_five_spend_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
sorted_df = renamed_top_five_spend_df.sort_values(by='Total Profit', ascending=False)

sorted_df
```
